```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import spearmanr

# Assuming CSS, SSS, FRR, and TRR are your dataframes

fig, ax = plt.subplots(2, 2, figsize=(20, 12))
ax = ax.flatten()

# Function to calculate and annotate Spearman correlation coefficient
def annotate_spearman_correlation(ax, x, y):
    corr, _ = spearmanr(x, y)
    ax.annotate(f'Spearman correlation: {corr:.2f}', xy=(0.5, 0.9), xycoords='axes fraction', ha='left', fontsize=12, fontweight='bold')

# Plotting ROOT ZONE SOIL WETNESS and PRECIPITATION for the first subplot
ax[0].plot(CSS.index, CSS.GWETROOT, color='green', label='ROOT ZONE SOIL WETNESS')
ax[0].set_xlabel('DAY OF THE YEAR')
ax[0].set_ylabel('ROOT ZONE SOIL WETNESS', color='green')
ax[0].tick_params(axis='y', labelcolor='green')
ax[0].grid(True, linestyle='--', linewidth=1, alpha=0.5, color='blue')

ax_twin = ax[0].twinx()
ax_twin.plot(CSS.index, CSS.PRECTOTCORR, color='blue', label='PRECIPITATION')
ax_twin.set_ylabel('PRECIPITATION (mm/day)', color='blue')
ax_twin.tick_params(axis='y', labelcolor='blue')

# Calculate and annotate Spearman correlation coefficient
annotate_spearman_correlation(ax[0], CSS.GWETROOT, CSS.PRECTOTCORR)

# Adding legend to the first subplot
lines, labels = ax[0].get_legend_handles_labels()
lines2, labels2 = ax_twin.get_legend_handles_labels()
ax[0].legend(lines + lines2, labels + labels2, loc='upper left')

# Setting title for the first subplot
ax[0].set_title('COASTAL SAVANNAH(CAPE COAST)', fontweight='bold', fontsize=17)

# Repeat the process for other subplots
# Plotting of ROOT ZONE SOIL WETNESS and PRECIPITATION for the second zone
ax[1].plot(SSS.index, SSS.GWETROOT, color='red', label='ROOT ZONE SOIL WETNESS')
ax[1].set_xlabel('DAY OF THE YEAR')
ax[1].set_ylabel('ROOT ZONE SOIL WETNESS', color='red')
ax[1].tick_params(axis='y', labelcolor='red')
ax[1].grid(True, linestyle='--', linewidth=1, alpha=0.5, color='blue')
ax_twin2 = ax[1].twinx()
ax_twin2.plot(SSS.index, SSS.PRECTOTCORR, color='blue', label='PRECIPITATION')
ax_twin2.set_ylabel('PRECIPITATION (mm/day)', color='blue')
ax_twin2.tick_params(axis='y', labelcolor='blue')


# Calculate and annotate Spearman correlation coefficient
annotate_spearman_correlation(ax[1], SSS.GWETROOT, SSS.PRECTOTCORR)


# Adding legend to the second subplot
lines, labels = ax[1].get_legend_handles_labels()
lines2, labels2 = ax_twin2.get_legend_handles_labels()
ax[1].legend(lines + lines2, labels + labels2, loc='upper left')

# Setting title for the second subplot
ax[1].set_title('SUDAN SAVANNAH(BAWKU)',fontweight='bold', fontsize=17)

#AGAIN

# Plotting ROOT ZONE SOIL WETNESS and PRECIPITATION for the first subplot
ax[2].plot(FRR.index, FRR.GWETROOT, color='red', label='ROOT ZONE SOIL WETNESS')
ax[2].set_xlabel('DAY OF THE YEAR')
ax[2].set_ylabel('ROOT ZONE SOIL WETNESS', color='red')
ax[2].tick_params(axis='y', labelcolor='red')
ax[2].grid(True, linestyle='--', linewidth=1, alpha=0.5, color='blue')
ax_twin = ax[2].twinx()
ax_twin.plot(FRR.index, FRR.PRECTOTCORR, color='blue', label='PRECIPITATION')
ax_twin.set_ylabel('PRECIPITATION (mm/day)', color='blue')
ax_twin.tick_params(axis='y', labelcolor='blue')


# Calculate and annotate Spearman correlation coefficient
annotate_spearman_correlation(ax[2], FRR.GWETROOT, FRR.PRECTOTCORR)

# Adding legend to the first subplot
lines, labels = ax[2].get_legend_handles_labels()
lines2, labels2 = ax_twin.get_legend_handles_labels()
ax[2].legend(lines + lines2, labels + labels2, loc='upper left')

# Setting title for the first subplot
ax[2].set_title('FOREST(KOFORIDUA)',fontweight='bold', fontsize=17)


# Plotting of ROOT ZONE SOIL WETNESS and PRECIPITATION for the second zone
ax[3].plot(TRR.index, TRR.GWETROOT, color='green', label='ROOT ZONE SOIL WETNESS')
ax[3].set_xlabel('DAY OF THE YEAR')
ax[3].set_ylabel('ROOT ZONE SOIL WETNESS', color='green')
ax[3].tick_params(axis='y', labelcolor='green')
ax[3].grid(True, linestyle='--', linewidth=1, alpha=0.5, color='blue')
ax_twin2 = ax[3].twinx()
ax_twin2.plot(TRR.index, TRR.PRECTOTCORR, color='blue', label='PRECIPITATION')
ax_twin2.set_ylabel('PRECIPITATION (mm/day)', color='blue')
ax_twin2.tick_params(axis='y', labelcolor='blue')

# Calculate and annotate Spearman correlation coefficient
annotate_spearman_correlation(ax[3], TRR.GWETROOT, TRR.PRECTOTCORR)


# Adding legend to the second subplot
lines, labels = ax[3].get_legend_handles_labels()
lines2, labels2 = ax_twin2.get_legend_handles_labels()
ax[3].legend(lines + lines2, labels + labels2, loc='upper left')

# Setting title for the second subplot
ax[3].set_title('TRANSITION(SUNYANI)',fontweight='bold', fontsize=17)


fig.subplots_adjust(top=0.85, hspace=0.5, wspace=0.3)  # Adjusting spacing between subplots

fig.suptitle('ROOT ZONE SOIL WETNESS AND PRECIPITATION VARIATION ACROSS ECOLOGICAL ZONES IN THE YEAR 2000', fontweight='bold', fontsize=20)
plt.savefig('TIME_SERIES.png', bbox_inches='tight', dpi=900)
plt.show()



```
